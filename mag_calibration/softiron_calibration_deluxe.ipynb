```python
'''
Calculating dataset of softiron values for the breadboard IMU
 '''

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

'''
Load and process the data
 '''

# Load the data
path = "/Users/haonan/OneDrive - Cornell University/research/ssds/acs_test_code/data/10.22.2024_voltimu_dataset/"

softirons = []

plt.figure(figsize=(6,4), dpi=200)
for i in np.arange(3.6, 4.2, .1): 
  print(i)
  data = pd.read_csv(path+'breadboard_voltimu_'+str(round(i,2))+'v.txt', header=1, index_col = False, names=['PWM', 'mag_x', 'mag_y', 'mag_z', 'offset_x', 'offset_y', 'offset_z', 'voltage'])

  pwm = data['PWM']
  mx = data['mag_x']
  my = data['mag_y']
  mz = data['mag_z']

  # Finding the pwm_index of when PWM is 0, but since this dataset specifically only has values for when PWM is -5 and 5, we are going to average it
  zero_pwm_index  = pwm[pwm == -5].index.tolist()[0]

  # mag readings when pwm==0
  mx_offset = (mx[zero_pwm_index] + mx[zero_pwm_index+1])/2
  my_offset = (my[zero_pwm_index] + my[zero_pwm_index+1])/2
  mz_offset = (mz[zero_pwm_index] + mz[zero_pwm_index+1])/2

  start = 0
  end = len(data)

  coeffs_mx = np.polyfit(pwm[start:end], mx[start:end]-mx_offset, 3)
  coeffs_my = np.polyfit(pwm[start:end], my[start:end]-my_offset, 3)
  coeffs_mz = np.polyfit(pwm[start:end], mz[start:end]-mz_offset, 3)

  softiron = [coeffs_mx[2], coeffs_mx[1], coeffs_mx[0], coeffs_my[2], coeffs_my[1], coeffs_my[0], coeffs_mz[2], coeffs_mz[1], coeffs_mz[0]]
  softirons.append(softiron)

# softirons = pd.DataFrame(softirons, columns=['3.6v', '3.7v', '3.8v', '3.9v', '4.0v', '4.1v', '4.2v'])
softirons = pd.DataFrame(softirons).T.rename(columns={0:'3.6v', 1:'3.7v', 2:'3.8v', 3:'3.9v', 4:'4.0v', 5:'4.1v', 6:'4.2v'})
softirons.to_csv('softiron_values_y.csv')
softirons
```

```python
''' 
Calculating dataset of softiron values for the cubesat IMU
'''

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

'''
Load and process the data
 '''

# Load the data
path = "/Users/haonan/OneDrive - Cornell University/research/ssds/acs_test_code/data/11.12.2024_voltimu_dataset/"

softirons = []

plt.figure(figsize=(6,4), dpi=200)
for i in np.arange(3.6, 4.2, .1): 
  print(i)
  data = pd.read_csv(path+'cubesat_ytorq_'+str(round(i,2))+'v.txt', names=['values'])

  pwm = []
  mx = []
  my = []
  mz = []

  for i in range(0,len(data), 6):
    pwm.append(float(data.loc[i+1]))
    mx.append(float(data.loc[i+2]))
    my.append(float(data.loc[i+3]))
    mz.append(float(data.loc[i+4]))

  pwm = pd.Series(pwm)
  mx = pd.Series(mx)
  my = pd.Series(my)
  mz = pd.Series(mz)

  # Finding the pwm_index of when PWM is 0, but since this dataset specifically only has values for when PWM is -5 and 5, we are going to average it
  zero_pwm_index  = pwm[pwm == -5].index.tolist()[0]

  # mag readings when pwm==0
  mx_offset = (mx[zero_pwm_index] + mx[zero_pwm_index+1])/2
  my_offset = (my[zero_pwm_index] + my[zero_pwm_index+1])/2
  mz_offset = (mz[zero_pwm_index] + mz[zero_pwm_index+1])/2

  start = 0
  end = len(data)

  coeffs_mx = np.polyfit(pwm[start:end], mx[start:end]-mx_offset, 3)
  coeffs_my = np.polyfit(pwm[start:end], my[start:end]-my_offset, 3)
  coeffs_mz = np.polyfit(pwm[start:end], mz[start:end]-mz_offset, 3)

  softiron = [coeffs_mx[2], coeffs_mx[1], coeffs_mx[0], coeffs_my[2], coeffs_my[1], coeffs_my[0], coeffs_mz[2], coeffs_mz[1], coeffs_mz[0]]
  softirons.append(softiron)

# softirons = pd.DataFrame(softirons, columns=['3.6v', '3.7v', '3.8v', '3.9v', '4.0v', '4.1v', '4.2v'])
softirons = pd.DataFrame(softirons).T.rename(columns={0:'3.6v', 1:'3.7v', 2:'3.8v', 3:'3.9v', 4:'4.0v', 5:'4.1v', 6:'4.2v'})
softirons.to_csv('softirons_y.csv')
softirons
```
