```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
# Load the data
path = "/Users/haonan/OneDrive - Cornell University/research/ssds/acs_test_code/data/10.09.2024_voltimu_dataset/"

mag_xs = []
mag_ys = []
mag_zs = []

plt.figure(figsize=(6,4), dpi=200)
for i in np.arange(3.6, 4.2, .1): 
  data = pd.read_csv(path+'voltimugraph_'+str(round(i,2))+'v.txt', header=1, index_col = False, names=['PWM', 'mag_x', 'mag_y', 'mag_z', 'offset_x', 'offset_y', 'offset_z', 'voltage'])
  mag_xs.append(data.mag_x)
  mag_ys.append(data.mag_y)
  mag_zs.append(data.mag_z)


# # data = data_4
pwm = data['PWM']

# pwmY_ox_1 = 8.83096680e-03;
# pwmY_ox_2 = 4.26409072e-07;
# pwmY_ox_3 = -6.69370023e-09;
# pwmY_oy_1 = -2.64514092e-01;
# pwmY_oy_2 = -9.82458813e-06;
# pwmY_oy_3 = 9.11136691e-08;
# pwmY_oz_1 = -1.90567242e-02;
# pwmY_oz_2 = -5.99945842e-06;
# pwmY_oz_3 = 7.85718685e-10;
# pwmY_ox = (pwmY_ox_1 * pwm) + (pwmY_ox_2 * pow(pwm, 2)) + (pwmY_ox_3 * pow(pwm, 3));
# pwmY_oy = (pwmY_oy_1 * pwm) + (pwmY_oy_2 * pow(pwm, 2)) + (pwmY_oy_3 * pow(pwm, 3));
# pwmY_oz = (pwmY_oz_1 * pwm) + (pwmY_oz_2 * pow(pwm, 2)) + (pwmY_oz_3 * pow(pwm, 3));

pwmY_ox_1 = 8.60013482e-03;
pwmY_ox_2 = -2.04542449e-06;
pwmY_ox_3 = 6.14475215e-09;
pwmY_oy_1 = -2.37677372e-01;
pwmY_oy_2 = -3.66292599e-07;
pwmY_oy_3 = 7.25220153e-08;
pwmY_oz_1 = -1.47995640e-02;
pwmY_oz_2 = 1.97761126e-07;
pwmY_oz_3 = 2.85273223e-08;
pwmY_ox = (pwmY_ox_1 * pwm) + (pwmY_ox_2 * pow(pwm, 2)) + (pwmY_ox_3 * pow(pwm, 3));
pwmY_oy = (pwmY_oy_1 * pwm) + (pwmY_oy_2 * pow(pwm, 2)) + (pwmY_oy_3 * pow(pwm, 3));
pwmY_oz = (pwmY_oz_1 * pwm) + (pwmY_oz_2 * pow(pwm, 2)) + (pwmY_oz_3 * pow(pwm, 3));

mx = (mag_xs[0] + mag_xs[1] + mag_xs[2] + mag_xs[3] + mag_xs[4] + mag_xs[5] + mag_xs[6])/7 - pwmY_ox
my = (mag_ys[0] + mag_ys[1] + mag_ys[2] + mag_ys[3] + mag_ys[4] + mag_ys[5] + mag_ys[6])/7 - pwmY_oy
mz = (mag_zs[0] + mag_zs[1] + mag_zs[2] + mag_zs[3] + mag_zs[4] + mag_zs[5] + mag_zs[6])/7 - pwmY_oz
# mx = (mag_xs[0] + mag_xs[1] + mag_xs[2] + mag_xs[3] + mag_xs[4] + mag_xs[5] + mag_xs[6])/7
# my = (mag_ys[0] + mag_ys[1] + mag_ys[2] + mag_ys[3] + mag_ys[4] + mag_ys[5] + mag_ys[6])/7
# mz = (mag_zs[0] + mag_zs[1] + mag_zs[2] + mag_zs[3] + mag_zs[4] + mag_zs[5] + mag_zs[6])/7


# mx = data['mag_x'] - pwmY_ox
# my = data['mag_y'] - pwmY_oy
# mz = data['mag_z'] - pwmY_oz
# mx = data['mag_x'] 
# my = data['mag_y'] 
# mz = data['mag_z'] 

# Finding the pwm_index of when PWM is 0, but since this dataset specifically only has values for when PWM is -5 and 5, we are going to average it
zero_pwm_index  = pwm[pwm == -5].index.tolist()[0]

# mag readings when pwm==0
mx_offset = (mx[zero_pwm_index] + mx[zero_pwm_index+1])/2
my_offset = (my[zero_pwm_index] + my[zero_pwm_index+1])/2
mz_offset = (mz[zero_pwm_index] + mz[zero_pwm_index+1])/2

start = 0
end = len(data)
```

```python
plt.figure(dpi=150)
plt.plot(pwm[start:end], mx[start:end], label = "mx")
plt.plot(pwm[start:end], my[start:end], label = "my")
plt.plot(pwm[start:end], mz[start:end], label = "mz")
# plt.plot(pd.Series(data_1.index), data_1['mag_y'], color='r', label='sep30_battery')
# plt.plot(pd.Series(data_2.index), data_2['mag_y'], color='r', label='sep30_battery')
# plt.plot(pd.Series(data_3.index), data_3['mag_y'], color='b', label='sep30_powersource')
# plt.plot(pd.Series(data_4.index), data_4['mag_y'], color='g', label='sep05_battery')
# plt.plot(pd.Series(data_5.index), data_5['mag_y'], color='orange', label='oct02_battery')
# plt.plot(pd.Series(data_6.index), data_6['mag_y'], color='orange', label='oct02_battery')
# plt.plot(pd.Series(data_7.index), data_7['mag_y'], color='orange', label='oct02_powersource')
# plt.plot(pd.Series(data_8.index), data_8['mag_y'], color='orange', label='oct02_battery')
# plt.plot(pd.Series(data_9.index), data_9['mag_y'], color='orange', label='oct02_powersource')
# plt.plot(pd.Series(data_10.index), data_10['mag_y'], color='orange', label='oct02_battery')
plt.xlabel("PWM")
plt.ylabel("uT")
plt.title("mag")
plt.legend(bbox_to_anchor=[1,1])
plt.show()

# plt.figure(figsize=(6,4), dpi=200)
# plt.plot(pd.Series(data.index), data.voltage)
```

```python
coeffs_mx = np.polyfit(pwm[start:end], mx[start:end]-mx_offset, 3)
coeffs_my = np.polyfit(pwm[start:end], my[start:end]-my_offset, 3)
coeffs_mz = np.polyfit(pwm[start:end], mz[start:end]-mz_offset, 3)

# Create polynomial functions from the coefficients
poly_mx = np.poly1d(coeffs_mx)
poly_my = np.poly1d(coeffs_my)
poly_mz = np.poly1d(coeffs_mz)
# Generate a range of PWM values for plotting the fit
pwm_fit = np.linspace(-255, 255, 500)

# Plot the original data
plt.figure(dpi=150)
plt.plot(pwm[start:end], mx[start:end]-mx_offset, '.', label="mx Data")
plt.plot(pwm[start:end], my[start:end]-my_offset, '.', label="my Data")
plt.plot(pwm[start:end], mz[start:end]-mz_offset, '.', label="mz Data")

# Plot the fitted curves
plt.plot(pwm_fit, poly_mx(pwm_fit), label="mx Fit")
plt.plot(pwm_fit, poly_my(pwm_fit), label="my Fit")
plt.plot(pwm_fit, poly_mz(pwm_fit), label="mz Fit")
plt.xlabel("PWM")
plt.ylabel("Magnetic Field (uT)")
plt.title("3rd Order Polynomial Fit of Magnetic Field vs. PWM")

plt.legend()
plt.show()

print("PWM_z to Magx: "+ str(coeffs_mx))
print("PWM_z to Magy: "+ str(coeffs_my))
print("PWM_z to Magz: "+ str(coeffs_mz))
```
